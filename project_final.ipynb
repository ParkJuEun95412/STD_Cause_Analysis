```python
from sklearn.svm import SVR

x = output.drop(['std_exp'] , axis = 1)
y = output['std_exp']

#  +,   
X_trainval, X_test, y_trainval, y_test = train_test_split(
    x,y, random_state=42)

# +  ,   
X_train, X_valid, y_train, y_valid = train_test_split(
    X_trainval, y_trainval, random_state=42)

print("  : {}       : {}       : {} \n".format(X_train.shape[0],
                                                       X_valid.shape[0], X_test.shape[0]))
```

```python
best_score = 0

for gamma in [0.001, 0.01, 0.1, 1, 10, 100]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        
        #     SVR 
        svm = SVR(gamma = gamma, C=C)
        svm.fit(X_train, y_train)
        
        #  SVC 
        score = svm.score(X_valid, y_valid)
        
        #      
        if score > best_score:
            best_score = score
            best_parameters = {'C':C, 'gamma':gamma}
```

```python
#             .

svm = SVR(**best_parameters)
svm.fit(X_trainval, y_trainval)
test_score = svm.score(X_test, y_test)
print("    : {:.2f}".format(best_score))
print(" :", best_parameters)
print("    : {:.2f}".format(test_score))
```
