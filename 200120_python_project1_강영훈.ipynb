```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model

X = output.drop(['std_exp'], axis=1)
y = output['std_exp']

linear_regression = linear_model.LinearRegression()
linear_regression.fit(X= pd.DataFrame(X), y =y) # 모델 생성
my_prediction = linear_regression.predict(X = pd.DataFrame(X)) # X 값을 넣어서 예측한 값 = predict

print('intercept = ', linear_regression.intercept_) # 절편
print('coef = ', linear_regression.coef_)           # 계수

print('---------------------------------------------------------')

# 적합도 검증 - 잔차
residuals = y - my_prediction
print(residuals.describe())


print('---------------------------------------------------------')
# 적합도 검증 - 결정계수 확인
SSE = (residuals**2).sum()
SST = ((y-y.mean())**2).sum()
R_squared = 1 - (SSE/SST)
print('R_squared = ', R_squared)

print('---------------------------------------------------------')
from sklearn.metrics import mean_squared_error

print('score = ', linear_regression.score(X = pd.DataFrame(X), y=y))
print('Mean_Squared_Error(MSE) = ', mean_squared_error(my_prediction, y)) # 평균제곱오차
print('RMSE = ', mean_squared_error(my_prediction, y)**0.5)              # 평균제곱근 편차

# 방정식 도출
# y = 0.49262012*X1 -0.04782063*X2 + 2.13626707*X3 -1.15083954*X4 -0.49214802*X5
#        -0.57899346*X6 -1.66715185*X7 + 0.18307615*X8 + 0.9245575916119094
```
