```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

pd.set_option('display.max_columns', None)
pr = pd.read_csv('prep_final.csv')
pr.index = pr["year"]
del (pr["year"])
pr
```

```python
pr.info()
```

```python
pr.isnull().sum() # 결측치 개수 확인
```

```python
df=pr.iloc[10:] # pr의 2010년 데이터부터 df에 담기.
df
```

```python
# 특정 부분 결측치 처리
df['contraception']=df['contraception'].fillna(method='ffill')
df['welfare_fac']=df['welfare_fac'].fillna(method='ffill')
df['sports_fac']=df['sports_fac'].fillna(method='ffill')
```

```python
df=df.fillna(df.mean())  # 나머지 부분 결측치 = 평균으로 대체
```

```python
df
```

```python
df.isnull().sum() # 결측치 개수 확인 -> 없어짐
```

```python
corr=df.corr(method='pearson')
corr # 상관계수
```

```python
from sklearn.preprocessing import StandardScaler
# 평균이 0, 표준편차가 1이 되도록 변환

trans_df = StandardScaler().fit_transform(df)
```

```python
trans_df
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2, svd_solver='full') # 변수 최대개수가 9개
principalComponents = pca.fit_transform(trans_df)

principalDf = pd.DataFrame(data = principalComponents)

print(principalDf)
print()

pca.fit(trans_df)

print(pca.explained_variance_ratio_) # [0.64112221 0.16125492]
# 새롭게 만든 축이 설명력을 얼마나 갖고 있는지 확인
# -> 첫 번째 변수가 더 설명력이 좋다.
```
